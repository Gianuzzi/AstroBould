```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("pp.dat", header=None, sep='\s+',
                   names=["t", "x", "y", "vx", "vy", "ax", "ay", "axn", "ayn"])
data["r"] = np.sqrt(data.x**2 + data.y**2)
data["v"] = np.sqrt(data.vx**2 + data.vy**2)
data["a"] = np.sqrt(data.ax**2 + data.ay**2)
data["an"] = np.sqrt(data.axn**2 + data.ayn**2)
data["vr"] = (data.x*data.vx + data.y*data.vy) / data.r
data["ar"] = (data.x*data.ax + data.y*data.ay) / data.r
data["arn"] = (data.x*data.axn + data.y*data.ayn) / data.r
data
```

```python
plt.plot(data.t, data.v, ".")
```

```python
plt.plot(data.t, data.vr, ".")
```

```python
fig, axs = plt.subplots(1,2, dpi=150)
axs[0].scatter(data.r, data.v, c=data.t)
p = axs[1].scatter(data.r, data.a, c=data.t)
fig.colorbar(p)
fig.tight_layout()
plt.show()
```
